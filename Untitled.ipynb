```python
CR=dp[(dp["cust_id"]==info["cust_id"][0])&(dp["debit_credit"]=='CR')]["tx_time"].unique()##第七筆貸方
DR=dp[(dp["cust_id"]==info["cust_id"][0])&(dp["debit_credit"]=='DR')]["tx_time"].unique()##第四筆借方
CUS=consume[consume["cust_id"]==info["cust_id"][0]]["date"].unique()
TRAN=trans[trans["cust_id"]==info["cust_id"][0]]["trans_date"].unique()
```

```python
## input=[]
## for id in info["cust_id"]: 下面 info....用id代替

mx=[[0]*12 for _ in range(395)]
for i in CR:
    mx[i][6]=dp[(dp["cust_id"]==info["cust_id"][0])&(dp["debit_credit"]=='CR')&(dp["tx_time"]==i)]["tx_amt"].sum()
    mx[i][7]=dp[(dp["cust_id"]==info["cust_id"][0])&(dp["debit_credit"]=='CR')&(dp["tx_time"]==i)]["tx_amt"].shape[0]
    mx[i][8]=dp[(dp["cust_id"]==info["cust_id"][0])&(dp["debit_credit"]=='CR')&(dp["tx_time"]==i)]["tx_amt"].max()
for i in DR:
    mx[i][3]=dp[(dp["cust_id"]==info["cust_id"][0])&(dp["debit_credit"]=='DR')&(dp["tx_time"]==i)]["tx_amt"].sum()
    mx[i][4]=dp[(dp["cust_id"]==info["cust_id"][0])&(dp["debit_credit"]=='DR')&(dp["tx_time"]==i)]["tx_amt"].shape[0]
    mx[i][5]=dp[(dp["cust_id"]==info["cust_id"][0])&(dp["debit_credit"]=='DR')&(dp["tx_time"]==i)]["tx_amt"].max()
for i in CUS:
    mx[i][0]=consume[(consume["cust_id"]==info["cust_id"][0])&(consume["date"]==i)]["amt"].sum()
    mx[i][1]=consume[(consume["cust_id"]==info["cust_id"][0])&(consume["date"]==i)]["amt"].shape[0]
    mx[i][2]=consume[(consume["cust_id"]==info["cust_id"][0])&(consume["date"]==i)]["amt"].max()
for i in TRAN :
    mx[i][9]=trans[(trans["cust_id"]==info["cust_id"][0])&(trans["trans_date"]==i)]["trade_amount_usd"].sum()
    mx[i][10]=trans[(trans["cust_id"]==info["cust_id"][0])&(trans["trans_date"]==i)]["trade_amount_usd"].shape[0]
    mx[i][11]=trans[(trans["cust_id"]==info["cust_id"][0])&(trans["trans_date"]==i)]["trade_amount_usd"].max()
### input.append(mx)    
```
