```python
temp=[]
for i in range(394):
    tempdata1=dp[(dp["cust_id"]==info["cust_id"][0])&(dp["debit_credit"]=='CR')&(dp["tx_time"]==i)]["tx_amt"]
    tempdata2=dp[(dp["cust_id"]==info["cust_id"][0])&(dp["debit_credit"]=='DR')&(dp["tx_time"]==i)]["tx_amt"]
    if tempdata1.empty and tempdata2.empty:
        temp.append([0,0,0,0,0,0])
    elif  tempdata1.empty:
        temp.append([0,0,0,sum(tempdata2),tempdata2.shape[0],max(tempdata2)])
    elif  tempdata2.empty:
        temp.append([sum(tempdata1),tempdata1.shape[0],max(tempdata1),0,0,0])
    else:
        temp.append([sum(tempdata1),tempdata1.shape[0],max(tempdata1),sum(tempdata2),tempdata2.shape[0],max(tempdata2)])
```
